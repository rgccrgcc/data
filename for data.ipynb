```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
import pylab as pl
```

```python
df1 = pd.read_csv('data.csv')
df1 = pd.DataFrame(df1.dropna())
```

```python
pl.figure()
pl.subplot(511)

df1.iloc[0:53]['1800'].plot(kind='bar')
pl.xticks(range(0,53), df1.iloc[0:53]['Country'].values)
pl.title('Subplot')
pl.xlabel('X')
pl.ylabel('Y')

pl.subplot(512)
df1.iloc[53:105]['1800'].plot(kind='bar')
pl.xticks(range(53,105),df1.loc[53:105]['Country'].values)
pl.title('Subplot')
pl.xlabel('X')
pl.ylabel('Y')

plt.subplots_adjust(top=15, right=4, left=1)
```
